```python
grupoMax = df.groupby(['Nombre_estacion','ID_estacion','dia'])['temt'].max()
grupoMin = df.groupby(['Nombre_estacion','ID_estacion','dia'])['temt'].min()
grupoProm = df.groupby(['Nombre_estacion','ID_estacion','dia'])['temt'].mean()
grupoMax = grupoMax.reset_index()
grupoMin = grupoMin.reset_index()
grupoProm = grupoProm.reset_index()
```

```python
grupoMax = grupoMax.rename(columns={'temt' : 'temt_max'})
grupoMin = grupoMin.rename(columns={'temt' : 'temt_min'})
grupoProm = grupoProm.rename(columns={'temt' : 'temt_prom'})
```

```python
nuevo = grupoMax.merge(grupoMin, on = ['Nombre_estacion','ID_estacion','dia'])
nuevo = grupoMin.merge(grupoProm, on = ['Nombre_estacion','ID_estacion','dia'])
```
