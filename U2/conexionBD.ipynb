```python
conn = pyodbc.connect(driver='{SQL Server}', server='(local)', database='ejemplo', trusted_connection='yes')
```

```python
cursor = conn.cursor()
```

```python
for index, rows in nuevo2.iterrows():
    insert = """INSERT INTO [dbo].[charts]
           ([Artista]
           ,[charts_spotify]
           ,[playlist_spotify]
           ,[charts_deezer]
           ,[playlist_deezer]
           ,[charts_apple]
           ,[playlist_apple])
     VALUES
           ('{a}',{b},{c},{d},{e},{f},{g})
           """.format(a= rows['Artista'],
                     b= rows['in_spotify_charts'], c= rows['in_spotify_playlists'],
                     d= rows['in_deezer_charts'], e= rows['in_deezer_playlists'],
                     f= rows['in_apple_charts'], g=rows['in_apple_playlists'])
    
    try:
        cursor.execute(insert)
    except:
        print(insert)
        break
conn.commit()
conn.close()
```
