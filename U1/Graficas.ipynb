```python
import pandas as pd
import matplotlib as plot
```

```python
df = pd.read_csv(r'C:\Users\caroc\OneDrive\Documentos\datos_youtube\datos1.csv') #la r se pone para dejar la 
```

```python
df
```

```python
df['fecha']=pd.to_datetime(df['fecha'])
```

```python
df['anio']=df['fecha'].dt.year
df['mes']=df['fecha'].dt.month
df['dia']=df['fecha'].dt.day
```

```python
df['hour']=df['fecha'].dt.hour
```

```python
df
```

```python
plotear = df[df.dia==1] #plotear es parte de un df 
```

```python
plotear.temt. plot.hist() #el plot.hist() es el formato de la grafica o bien, la forma en que se vera 
```

```python
plotear.plot.scatter(x="dirv", y="velv", alpha=0.2)
```

```python
temt_prom = plotear.groupby("hour")["temt"].mean()
temt_prom.head(25).plot.barh()
```

```python
plotear.velv.value_counts().plot.pie()
```
